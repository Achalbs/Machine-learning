```python
# Importing Necessary Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

1. Loading and Preprocessing


```python
df = pd.read_csv('CarPrice_Assignment.csv')
```

```python
df.tail(10)
```

Extract brand from CarName


```python
df['CarBrand'] = df['CarName'].apply(lambda x: x.split()[0].lower())
```

```python
df.tail(10)
```

```python
# Fix misspellings
df['CarBrand'] = df['CarBrand'].replace({'vw': 'volkswagen', 'vokswagen': 'volkswagen', 'maxda': 'mazda', 'porcshce': 'porsche', 'toyouta': 'toyota'})
```

```python
# Drop unnecessary columns
df.drop(['car_ID', 'CarName'], axis=1, inplace=True)
```

one-hot encoding on all categorical variables in the DataFrame.



```python
# Convert categorical to dummies
df = pd.get_dummies(df, drop_first=True)
```

Separate features and target .



```python
X = df.drop('price', axis=1)
y = df['price']
```

Scaling features



```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

Test_Train_Split



```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
2. Model Implementation
```

```python
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree": DecisionTreeRegressor(random_state=42),
    "Random Forest": RandomForestRegressor(random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(random_state=42),
    "SVR": SVR()
}

results = {}
predictions = {}
```

3. Model Evaluation


```python
for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    predictions[name] = y_pred
    results[name] = {
        'R2 Score': r2_score(y_test, y_pred),
        'MSE': mean_squared_error(y_test, y_pred),
        'MAE': mean_absolute_error(y_test, y_pred)
    }
```

```python
results_df = pd.DataFrame(results).T
```

Model Evaluation Metrics:


```python
print(results_df)
```

Plotting Predictions vs Actual price


```python
# Plot Predictions vs Actuals
for name in predictions:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x=y_test, y=predictions[name])
    plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], '--r')
    plt.title(f'{name} - Predicted vs Actual Prices')
    plt.xlabel('Actual Price')
    plt.ylabel('Predicted Price')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

4. Feature Importance (for tree-based models)


```python
rf_model = models["Random Forest"]
importances = pd.Series(rf_model.feature_importances_, index=X.columns).sort_values(ascending=False)
```

Top 10 Important Features :


```python
print(importances.head(10))
```

Plotting feature importances .


```python
importances.head(10).plot(kind='barh', title='Top 10 Important Features')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

5. Hyperparameter Tuning (for best model, say Random Forest)(for tree-based models)


```python
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, 15, 20],
    'min_samples_split': [2, 5],
}
```

```python
grid_search = GridSearchCV(RandomForestRegressor(random_state=42), param_grid, cv=5, scoring='r2', n_jobs=-1)
grid_search.fit(X_train, y_train)
best_model = grid_search.best_estimator_
y_pred_tuned = best_model.predict(X_test)
```

Tuned Random Forest Performance:.


```python
print("R2 Score:", r2_score(y_test, y_pred_tuned))
print("MSE:", mean_squared_error(y_test, y_pred_tuned))
print("MAE:", mean_absolute_error(y_test, y_pred_tuned))
```
